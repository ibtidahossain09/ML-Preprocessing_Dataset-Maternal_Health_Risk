```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/pre processing /Maternal_Health_Risk.csv")

# ----------------------------
# Pairplot for Original Data
# ----------------------------
sns.pairplot(df, hue='RiskLevel')
plt.suptitle("Original Maternal Health Risk Data", y=1.02)
plt.show()

# ----------------------------
# Min-Max Normalized Data
# ----------------------------
df_minmax = df.copy()
features = df_minmax.drop('RiskLevel', axis=1)
scaler = MinMaxScaler()
df_minmax[features.columns] = scaler.fit_transform(features)

sns.pairplot(df_minmax, hue='RiskLevel')
plt.suptitle("Min-Max Normalized Maternal Health Data", y=1.02)
plt.show()

# ----------------------------
# Z-score Normalized Data
# ----------------------------
df_zscore = df.copy()
features = df_zscore.drop('RiskLevel', axis=1)
scaler = StandardScaler()
df_zscore[features.columns] = scaler.fit_transform(features)

sns.pairplot(df_zscore, hue='RiskLevel')
plt.suptitle("Z-score Normalized Maternal Health Data", y=1.02)
plt.show()
```
